# Data Visualizations

## Matplotlib

```python
# Line Plot

import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
#plt.plot(year, pop, color="blue")
plt.plot(year, pop, color="y")
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('World Population in Line Plot')
plt.show()
```

```python
# Scatter Plot

import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.scatter(year, pop, color="b")
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('World Population in Scatter Plot')
plt.show()
```

```python
# Histogram

values = [0,0.6, 1.4, 1.6, 2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
plt.hist(values, bins=3)
plt.title('histogram bins')
plt.hist(values, color='r')
plt.show()
```
